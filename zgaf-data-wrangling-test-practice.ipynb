```python
import pandas as pd
```

```python
# 1.- Cargar el archivo a un Data Frame llamado “cereal” 
cereal = pd.read_csv('./files/cereals.csv')
```

```python
cereal
```

```python
# 1.- Crear un subconjunto de 20 registros completos (independiente llamado “scereal” con los atributos (name, Manuf, Type, Calories, Protein, Sodium, Potass, Vitamins).
scereal = cereal[['Name', 'Manuf', 'Type', 'Calories', 'Protein', 'Sodium', 'Potass', 'Vitamins']].tail(20)
```

```python
# 2.- Calcular en “scereal” la raíz cuadrada, media, desviación estandar de 'Calories' 
import numpy as np

scereal.agg({'Calories':['median', 'std']})
np.sqrt(scereal[['Calories']].apply(sum))

```

```python
# 2.- Calcular en “scereal” la raíz cuadrada, media, desviación estandar de 'Sodium' 
scereal.agg({'Sodium':['median', 'std']})
np.sqrt(scereal[['Sodium']].apply(sum))
```

```python
# 3.- Sustituir los valores “0” por la media del atributo “Sodium” en “cereal”.
cereal['Sodium'] = cereal['Sodium'].replace(0, cereal['Sodium'].mean())
```

```python
# 4.- Eliminar los valores “NA” (nulos) en “cereal” y asignarlos a una variable llamada “sncereal”.
sncereal = cereal.dropna()
```

```python
# 6.- Filtrar "cereal", los atributos (Manuf = K) y (Potass >= 150)
cereal[(cereal['Manuf'] == 'K ') & (cereal['Potass']>=150)]
```

```python
# 7.- Filtrar "scereal" por grupos de acuerdo a los atributos (name,Type) y sumarizar el campo ("Calories"), ordenarlo de mayor a menor (descendente)
scereal.groupby(['Name', 'Type'])[['Calories']].apply(sum).sort_values('Calories', ascending = False)
```

```python
# 8.- Tomando como base "scereal", encontrar las funciones ("media", "desviación estandar", "raíz cuadrada") que se aplique a todos los campos cuantitativos (es decir, Calories, Protein, Sodium, Potass, Vitamins). Encontrar una forma elegante de resumirlo (existen varias alternativas).
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
scereal.select_dtypes(include=numerics).agg(['min', 'max', 'mean', 'median', 'std'])
```

```python
# 9.- Utilizando el archivo "cereals2.csv", combinar el conjunto de datos "cereals.csv" con "cereals2.csv" tomando como base el atrinuto "name" y nombrarlo "supercereal"
supercereal = pd.merge(cereal, pd.read_csv('./files/cereals2.csv'), on="Name") 
supercereal
```

```python
# 10.- Utilizar el Data Frame "supercereal" y aplicar las funciones descritas en el punto 8.
supercereal.select_dtypes(include=numerics).agg(['min', 'max', 'mean', 'median', 'std'])
```
