```python
# 1.- Cargar el archivo a un Data Frame llamado “bk”
bk = pd.read_csv('./files/bank2.csv')
bk
```

```python
# 1.- Crear un subconjunto de 50 registros completos (independiente llamado “sbk” con los atributos (age, job, education, marital, month,balance, campaign).
sbk = bk[['age', 'job', 'education', 'marital', 'month', 'balance', 'campaign']].head(50)
sbk
```

```python
# 2.- Calcular en “sbk” la media y desviación estandar de los campos (age, balance).
sbk.agg({'age':['median', 'std'], 'balance':['median', 'std']})
```

```python
# 3.- Sustituir los valores “0” por la media del atributo “balance” en “bk”.
bk['balance'] = bk['balance'].replace(0, bk['balance'].mean())
bk
```

```python
# 4.- Crear una tabla de frecuencias (conteo) para los atributos (“job”, “education”) y asignarla a una variable “tbk” del dataset “sbk”.
tbk = pd.DataFrame(sbk[['job', 'education']].value_counts())
#tbk = sbk.groupby(['job', 'education']).size().reset_index(name='count')
tbk
```

```python
# 5.- Sustituir en el atributo “education” los valores “unknown” por “primary” y asignarlo a una variable llamada “edu” del dataset “bk”.
edu = pd.DataFrame(bk['education'].replace('unknown', 'primary'))
edu 
```

```python
# 6.- Filtrar "bk", los atributos (education = “secondary”) y (loan = “no") o (duration >= 100)
bk[(bk['education'] == 'secondary') & (bk['loan']=='no') | (bk['duration'] >= 100)]
```

```python
# 7.- Filtrar "bk" por grupos de acuerdo a los atributos (“job”, “contact”) y sumarizar el campo ("balance"), ordenarlo de mayor a menor (descendente)
bk.groupby(['job', 'contact'])[['balance']].apply(sum).sort_values('balance', ascending = False)
```

```python
# 8.- Obtener una tabla cruzada donde pueda obtener la media (mean) del atributo “age” y la desviación estandar del atributo “balance” agrupadas por el atributo “education”.
round(pd.DataFrame(bk.groupby(['education'], as_index=True).agg({'age':'mean', 'balance':'std'})),2)
```
